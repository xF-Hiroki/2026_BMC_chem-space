```python
import matplotlib.pyplot as plt

# データ
total_cpd = 79289
selective_hits = 273
unselective_hits = 34

# ヒット率（%）
unselective_hit_rate = unselective_hits / total_cpd * 100
selective_hit_rate = selective_hits / total_cpd * 100

# ラベルとデータ
categories = ['selective', 'unselective']
hit_rates = [selective_hit_rate, unselective_hit_rate]

# カスタムカラーパレット
palette = {
    "selective": "Aquamarine",
    "unselective": "LightSeaGreen"
}
colors = [palette[cat] for cat in categories]

# グラフ描画
fig, ax = plt.subplots(figsize=(4, 6))
bars = ax.bar(categories, hit_rates, color=colors, edgecolor='black')


# 軸ラベル・目盛のスタイル調整
ax.set_ylabel('Hit Rate (%)', fontsize=24)
ax.tick_params(axis='both', labelsize=20)

# グリッド非表示
ax.grid(False)

plt.tight_layout()
plt.show()
```
