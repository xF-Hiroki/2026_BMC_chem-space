```python
import forgi
def dotbraket2element(structure):
    fg = forgi.load_rna(structure, allow_many=False)
    return fg.to_element_string(with_numbers=True).split('\n')[0]

def dotbraket2element_num(structure):
    fg = forgi.load_rna(structure, allow_many=False)
    return 'elem_num' + str(fg.to_element_string(with_numbers=True).split('\n')[1])
```

```python
import pandas as pd
data = pd.read_csv('./data/0_input/rna_library.csv')
data['RNAProbeStructurePredict_element'] = data['RNAProbeStructurePredict_Step1'].apply(dotbraket2element)
data['RNAProbeStructurePredict_element_num'] = data['RNAProbeStructurePredict_Step1'].apply(dotbraket2element_num)
data
```

```python
data.to_csv('./data/0_input/input_for_KNIME_rna.csv')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = data
```

```python
df.head()
```

```python

# === label size ===
label_size = 20
title_size = 20
tick_size  = 18

# === base rate ===
def calc_base_ratios(seq):
    seq = seq.upper()
    length = len(seq)
    return {
        'A': seq.count('A') / length if length else 0,
        'U': seq.count('U') / length if length else 0,
        'C': seq.count('C') / length if length else 0,
        'G': seq.count('G') / length if length else 0
    }

# === 比率の計算と整形 ===
base_ratio_df = df['Sequence'].apply(calc_base_ratios).apply(pd.Series)
base_ratio_df['id'] = df.index  # 元のインデックス保持

# meltでlong形式に変換（violinplotに便利）
melted = base_ratio_df.melt(id_vars='id', var_name='Base', value_name='Ratio')

# === Violin plot 描画 ===
plt.figure(figsize=(8, 6))
sns.violinplot(
    x='Base', y='Ratio', data=melted,
    color='turquoise'
)
plt.title("Base Composition per RNA", fontsize=title_size)
plt.xlabel("Base", fontsize=label_size)
plt.ylabel("Ratio", fontsize=label_size)
plt.xticks(fontsize=tick_size)
plt.yticks(fontsize=tick_size)
plt.tight_layout()
plt.show()

```
